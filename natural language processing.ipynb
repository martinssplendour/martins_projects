```python
import nltk
import matplotlib.pyplot as plt
nltk.download("punkt")
from nltk.tokenize import sent_tokenize
```

```python
sentences = sent_tokenize(text)
```

```python
sentences[2]
```

```python
import re

text = ' '.join(re.sub(r'[^A-Za-z0-9\s]', '', sentence) for sentence in sentences)
```

# TOKENIZATION

```python
from nltk.tokenize import word_tokenize
```

```python
words = word_tokenize(text)
```

# Removal of stop words

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
words = [w for w in words if w not in stopwords.words("english")]
```

# Stemming and lemmatization

```python
nltk.download('wordnet') #download for lemmatization
nltk.download('omw-1.4')
```

```python
from nltk.stem.porter import PorterStemmer

stemmed = [PorterStemmer().stem(w) for w in words]
```

```python
from nltk.stem.wordnet import WordNetLemmatizer

lemmatized = [WordNetLemmatizer().lemmatize(w) for w in words]
```

```python
text_emotion = []
with open("emotions.txt", 'r') as emotions:
    for line in emotions:
        clearline = line.replace('\n', '').replace(',', '').replace("'", '').strip()
        wording, emotion = clearline.split(':')
        if wording in words:
            text_emotion.append(emotion)

print(text_emotion)
```

```python
# Step 2: Preprocess the data by combining features
movies_df['Combined'] = (
    movies_df['Description'] + " " + 
    movies_df['Genres'] + " " + 
    movies_df['Actors'] + " " +
    movies_df['Ratings'].astype(str)
)

# Step 3: Vectorize the combined features using TF-IDF
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(movies_df['Combined'])
```

```python
# Step 4: Compute similarity scores using cosine similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

cosine_sim
```

```python
# Step 5: Recommendation function
def recommend_movies(title, cosine_sim=cosine_sim):
    # Get the index of the selected movie
    idx = movies_df[movies_df['Title'] == title].index[0]
    
    # Get similarity scores for all movies
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # Sort by similarity scores in descending order
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Get the top 3 most similar movies (excluding the selected movie itself)
    sim_scores = sim_scores[1:7]
    movie_indices = [i[0] for i in sim_scores]
    
    return movies_df['Title'].iloc[movie_indices]
```

```python
# Step 6: Test the recommendation system
print("Recommendations for 'The Empire Strikes Back':")
print(recommend_movies('The Empire Strikes Back'))
```
